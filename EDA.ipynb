```python
# funcion de conversion de las columnas anidadas : genres,production_companies ,production_countries ,spoken language ,cast 

import ast
def conversion_listas_Dict(columna): 
    o=[]
    for i in range(len(columna)):
        p=[]
        for j in range(len(ast.literal_eval(columna[i]))): 
            p.append(ast.literal_eval(columna[i])[j]['name'])
        o.append(p) # lista de listas
    ser1=pd.Series(o)
    return ser1
   
```

```python
# para la columna crew

def conversion_listas_Dicc(columna): 
    o=[]
    for i in range(len(columna)):
        p=[]
        for j in range(len(ast.literal_eval(columna[i]))): 
            if ast.literal_eval(columna[i])[j]['job']=='Director':
               p.append(ast.literal_eval(columna[i])[j]['name'])
        o.append(p) 
    serie=pd.Series(o)
    return serie
```

```python
# conversion de las columnas cast , genres,production_companies ,production_countries ,spoken language a listas

df_muestra['cast']=conversion_listas_Dict(df_muestra['cast'])
df_muestra['genres']=conversion_listas_Dict(df_muestra['genres'])
df_muestra['production_companies']=conversion_listas_Dict(df_muestra['production_companies'])
df_muestra['production_countries']=conversion_listas_Dict(df_muestra['production_countries'])
df_muestra['spoken_languages']=conversion_listas_Dict(df_muestra['spoken_languages'])
```

```python
#conversion de la columna crew
df_muestra['crew']=conversion_listas_Dicc(df_muestra['crew'])
```

```python
# datos atipicos encontrados 
def datos_atipicos(columna):
    atipicos4 = []
    columna1 = columna.sort_values(ascending=True)
    q1 = columna1.quantile(0.25)
    q3 =columna1.quantile(0.75)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    dtos4=list(columna1)
    for i in dtos4: 
        if (i<lwr_bound or i>upr_bound):
            atipicos4.append(i)
    return {'nombre_de_la_variable' :columna.name,'cantidad de atipicos': len(atipicos4) }
```

```python
#cantidad de datos atipicos u outliers en las variables numericas mencionadas .

datos_atipicos_en_popularidad=datos_atipicos(df_muestra['popularity'])
datos_atipicos_en_runtime=datos_atipicos(df_muestra['runtime'])
datos_atipicos_en_revenue=datos_atipicos(df_muestra['revenue'])
datos_atipicos_en_budget=datos_atipicos(df_muestra['budget'])
print(datos_atipicos_en_popularidad)
print(datos_atipicos_en_runtime)
print(datos_atipicos_en_revenue)
print(datos_atipicos_en_budget)
```
